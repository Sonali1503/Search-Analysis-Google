```python
!pip install pytrends
!pip install matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pytrends.request import TrendReq
```

Here, we will import all the required libraries to start our analysis.


```python
trends = TrendReq()
n=trends.build_payload(kw_list=["Machine Learning"])
data=trends.interest_by_region()
data=data.sort_values(by="Machine Learning", ascending=False)
data=data.head(10)
print(data)
```

The above table shows the number of searches for 'Machine  Learning' for each country.

```python
data.reset_index().plot(x="geoName", y="Machine Learning", figsize=(10,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

Above, we have plotted a bar graph for Number or Searches v/s Country name. We have set the x-axis name as geoName.

```python
data=TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data=data.interest_over_time()
fig, ax = plt.subplots(figsize=(20,15))
data['Machine Learning'].plot
plt.style.use('fivethirtyeight')
plt.title('Total Searches for Machine Learning', fontweight='bold')
plt.xlabel("Year")
plt.ylabel("No. of Searches")
plt.show()
```
